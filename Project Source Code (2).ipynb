# BarPlot Of Male,Female & Transgender Below 18 years

```python
df1 = {'Cause':df.Cause,
        'Below_18_years_Male': df.Below_18_years_Male,
        'Below_18_years_Female': df.Below_18_years_Female,
        'Below_18_years_Transgender': df.Below_18_years_Transgender}
df2= pd.DataFrame(df1)
```

```python
df2
```

```python
sns.barplot(data=df2)
plt.title('Bar Plot of Below 18 Years by Gender')
plt.xlabel('Gender')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df2 is your DataFrame containing the data

plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Combine data for all genders into a single DataFrame
df_melted = df2.melt(id_vars='Cause', 
                     value_vars=['Below_18_years_Male', 'Below_18_years_Female', 'Below_18_years_Transgender'], 
                     var_name='Gender', value_name='Count')

# Create the bar plot with hue
sns.barplot(x='Cause', y='Count', hue='Gender', data=df_melted)

plt.title('Bar Plot of Below 18 Years by Gender')
plt.xlabel('Cause')
plt.ylabel('Count')
plt.legend(title='Gender',loc=1)
plt.xticks(rotation=90)
plt.tight_layout()  # Ensures proper spacing of elements
plt.show()
```

# In The Above Graph We Analyze :-
## That Male Count Is Suiciding More Due To Family Problems
## & Females Are Also Suiciding More In Issue Of Family Problems
## The Transgender Are Suiciding Due To illness
## Overall It Is Observed That Most Suicides Are Happening Because Of Family Problems
